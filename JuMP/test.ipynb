```julia
using JuMP, GLPK, Ipopt, Cbc
```

```julia
m= Model(Cbc.Optimizer)
```

```julia
#varinfo(JuMP)
```

```julia
@variable(m,x>=0)
@variable(m,y>=0)
```

```julia
@constraint(m,x+y<=30)
@constraint(m,x-y>=3)
@constraint(m,5*x+4*y <= 35)
```

```julia
@objective(m,Max,5x+6y)
```

```julia
@show m
```

```julia
optimize!(m)
```

```julia
termination_status(m)
```
